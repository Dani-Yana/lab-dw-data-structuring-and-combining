```python
# Your code goes here

import pandas as pd
import numpy as np
```

```python
file1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
file2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
file3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')

file1.head()
```

```python
file2.head()
```

```python
file3.head()
```

```python
file1.shape
```

```python
file2.shape
```

```python
file3.shape # all 3 files have the same number of columns 
```

```python
# change column names so that they`re all the same
# first converting all column names to lower case

file_list = [pd.DataFrame(file1), pd.DataFrame(file2), pd.DataFrame(file3)]

for df in file_list:
    df.columns = df.columns.str.lower()

#for file_cols in file_list:
    #file_list.file_cols = file_list.file_cols.str.lower()
    
#    Questions: how does it know that columns are actually the colunmns?!?

# all state columns same name

for df in file_list:
    df = df.rename(columns={'st': 'state'}, inplace=True) # without inplace=True the output would be a DF with the new column names

for df in file_list:
    df.columns = df.columns.str.replace(' ', '_') # same question as above
    
```

```python
# now that all columns have the same headers we can proceed to combine all 3 tables: 

customer_data = pd.concat([file1, file2, file3], axis=0)

customer_data.shape
# the number of the rows doesn`t correspond the sum of the rows of the 3 files :( 
```

```python
customer_data.reset_index(drop=True, inplace=True)

customer_data.shape
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here

data = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')
display(data)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

# 2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Solution to 1.

rev_per_channel=data.pivot_table(index='sales_channel', values='total_claim_amount').round(2)
display(rev_per_channel)
rev_per_channel.describe()
```

```python
# Conclusion: the total claim amount is pretty evenly distributed between the different sales channels. 
```

```python
# Solution to 2: 

acv_gender_edu=data.pivot_table(index='education', columns='gender', values='customer_lifetime_value').round(2)

display(acv_gender_edu)
acv_gender_edu.describe()
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table. 

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Your code goes here

# There is no column with information about number of complaints in the given dataframe. Alternatively 
# I used total_claim_amount

no_compl_per_month_by_poltype=data.pivot_table(index='policy_type', columns='month', values='total_claim_amount').round(2)
display(no_compl_per_month_by_poltype)
```
